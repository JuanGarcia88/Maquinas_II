```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def magnetization_curve(If):
    if 0.0 < If <= 7.6:
        return 49.57 * If + 15.7
    elif 7.6 < If <= 10:
        return 18.38* If + 252.75



If_values = np.linspace(0.0, 10.0, 300)
Ea_values = np.array([magnetization_curve(If) for If in If_values])
idx = np.abs(If_values - 7.55).argmin()
crossing_Ea = Ea_values[idx]
magnetization_values = np.vectorize(magnetization_curve)(If_values)
plt.figure(figsize=(8, 6))
plt.plot(If_values, magnetization_values, label='Curva de Magnetización')
plt.plot(7.55, crossing_Ea, 'ro', label='Punto de Cruce (If = 7.55)')
plt.xlabel('Corriente (If)')
plt.ylabel('Magnetización')
plt.title('Curva de Magnetización')
print(f'El valor de Ea en el punto de cruce (If = 7.55) es: {crossing_Ea}')
plt.legend()
plt.grid(True)
plt.show()
```
